### Feature Engineering Process

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(5)
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
final_df.head(5)
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrial_Time_Hour']=final_df['Arrival_Time'].apply(lambda x: x.split(':')[0])
final_df['Arrial_Time_Min']=final_df['Arrival_Time'].apply(lambda x: x.split(':')[1])
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df['Arrial_Time_Hour']=final_df['Arrial_Time_Hour'].astype(int)
final_df['Arrial_Time_Min']=final_df['Arrial_Time_Min'].astype(int)

```

```python
final_df['Dep_Time_Hour']=final_df['Dep_Time'].apply(lambda x: x.split(':')[0])
final_df['Dep_Time_Min']=final_df['Dep_Time'].apply(lambda x: x.split(':')[1])
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df['Dep_Time_Hour']=final_df['Dep_Time_Hour'].astype(int)
final_df['Dep_Time_Min']=final_df['Dep_Time_Min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.head(3)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.head(5)
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head(3)
```

```python
final_df['Durattion_Hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Durattion_Hour']=final_df['Durattion_Hour'].astype('int')
```

```python
final_df['Durattion_Min']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df['Durattion_Hour'] = final_df['Durattion_Hour'].apply(lambda x: x*60)
```

```python
final_df['Durattion_Min'].isnull().sum()
```

```python
final_df['Durattion_Min'].unique()
```

```python
final_df['Durattion_Min'].fillna(final_df['Durattion_Min'].mode()[0], inplace=True)
```

```python
final_df['Durattion_Min'].isnull().sum()
```

```python
final_df['Durattion_Min'].unique()
```

```python
final_df['Durattion_Min']=final_df['Durattion_Min'].astype('int')
```

```python
sum_column = final_df['Durattion_Hour']+ final_df['Durattion_Min']
final_df["Total_Duration"] = sum_column
```

```python
final_df.drop('Duration',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder=LabelEncoder()
```

```python
final_df['Airline']=labelEncoder.fit_transform(final_df['Airline'])
final_df['Source']=labelEncoder.fit_transform(final_df['Source'])
final_df['Destination']=labelEncoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelEncoder.fit_transform(final_df['Additional_Info'])
```
