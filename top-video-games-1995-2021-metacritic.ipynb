```python
dfvalued=data.groupby("platform", as_index=False).meta_score.mean()
print(" top 10 most meat score of platform")
dfvalued.sort_values(by="meta_score",ascending=False).head(10)
```

```python
dfvalued=data.groupby(['name','platform','user_review'],as_index=False).meta_score.max()
print("Highest review game in each Platform")
dfvalued.sort_values(by="meta_score",ascending=False).head(10)
```

**movie released from 2000-2001**

```python
data["year"] = [x.split(",")[1] for x in data["release_date"].values]
```

```python
df = pd.DataFrame(data)
start_date = '2000'
end_date = '2021'
mask = (df['year'] > start_date) & (df['year'] <= end_date)
df = df.loc[mask]
display(df)
```
